    .distances: list[list]

```python
import netket,numpy
```

```python
sz=[[1,0],[0,-1]]
sx=[[0,1],[1,0]]
sy=[[0,-1j],[1j,0]]
graph=netket.graph.CustomGraph([[0,1,0],])
hilbert=netket.hilbert.Spin(s=0.5,graph=graph)
hamiltonian=netket.operator.GraphOperator(hilbert
                ,bondops=[(-1*numpy.kron(sz,sz)).tolist()]
                ,bondops_colors=[0,])
```

```python
ma=netket.machine.RbmSpin(alpha=1,hilbert=hamiltonian.hilbert)
#ma.init_random_parameters(sigma=0.01)
ma.parameters=[2,0,0,0,0,0,0,0]
print(ma.parameters)
print(ma.to_array())
ma.log_val([1,1,0,0,0])
```
